```python
# drop the name out of the data set
X = df.drop('name', axis=1)
```

```python
# perform clustering with 4 clusters
song_cluster = KMeans(n_clusters=4, init='random')
song_cluster.fit(X)
y_kmeans = song_cluster.predict(df.drop('name', axis=1))
df['cluster'] = y_kmeans
```

```python
song_cluster.labels_
```

```python
################
### Exercise ###
################

# List the songs in each cluster



df['cluster'] = song_cluster.labels_
df.sort('cluster')
```

```python

# save the DataFrame of cluster centers
centers = df.groupby('cluster').mean()

# calculate the mean of each feature for each cluster
df.groupby('cluster').mean()
```
